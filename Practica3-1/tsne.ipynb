```python
%matplotlib inline
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
sns.set()
sns.set(rc={"figure.figsize": (10, 8)})
PALETTE = sns.color_palette('deep', n_colors=3)
CMAP = ListedColormap(PALETTE.as_hex())
RANDOM_STATE = 42
```

```python
dataset = load_iris()
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
target = 'species'
iris = pd.DataFrame(
    dataset.data,
    columns=features)
iris[target] = dataset.target
iris.head()
```

```python
import pandas as pd

# Cargar el conjunto de datos Iris
iris_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
iris_data = pd.read_csv(iris_url, header=None, names=column_names)

# Calcular la media de cada clase
means_by_class = iris_data.groupby("class").mean()

# Mostrar los resultados
print("Media de cada clase en el conjunto de datos Iris:")
print(means_by_class)
```

```python
def plot_iris_2d(x, y, title, xlabel="x", ylabel="y"):
    sns.set_style("darkgrid")
    
    plt.scatter(x, y,
        c=iris['species'],
        cmap=CMAP,
        s=70)
    
    plt.title(title, fontsize=20, y=1.03)
    plt.xlabel(xlabel, fontsize=16)
    plt.ylabel(ylabel, fontsize=16)
    plt.show()  

def plot_iris_3d(x, y, z, title):
    sns.set_style('whitegrid')
    
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    
    ax.scatter(x, y, z,
               c=iris['species'],
               cmap=CMAP,
               s=40)
    
    ax.set_title(title, fontsize=20, y=1.03)
    
    fsize = 14
    ax.set_xlabel("x", fontsize=fsize)
    ax.set_ylabel("y", fontsize=fsize)
    ax.set_zlabel("z", fontsize=fsize)
    
    ax.xaxis.set_ticklabels([])
    ax.yaxis.set_ticklabels([])
    ax.zaxis.set_ticklabels([])
    plt.show() 
```

```python
tsne = TSNE(n_components=2, n_iter=1000, random_state=RANDOM_STATE)
points = tsne.fit_transform(iris[features])
plot_iris_2d(
    x = points[:, 0],
    y = points[:, 1],
    title = 'Reducción con t-SNE')
```

```python
tsne = TSNE(n_components=3, n_iter=5000, random_state=RANDOM_STATE)
points = tsne.fit_transform(iris[features])
plot_iris_3d(
    x=points[:, 0],
    y=points[:, 1],
    z=points[:, 2],
    title="Reducción con t-SNE en 3D")
```

```python
#-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-SANTA ORDENACION-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*

lar_set=[]
an_set=[]

lar_ver=[]
an_ver=[]

lar_vir=[]
an_vir=[]

c1=[]
c2=[]

tsne = TSNE(n_components=2, n_iter=1000, random_state=RANDOM_STATE)
X_tsne2 = tsne.fit_transform(iris[features])

# Matrices por separado
for i in range(len(X_tsne2)):
  c1.append(X_tsne2[i][0])
  c2.append(X_tsne2[i][1])

cont=0

for i in c1:
  if cont<50:
    lar_set.append(i)
  if cont>=50 and cont<100:
    lar_ver.append(i)
  if cont>=100 and cont<150:
    lar_vir.append(i)
  cont+=1

cont=0
for i in c2:
  if cont<50:
    an_set.append(i)
  if cont>=50 and cont<100:
    an_ver.append(i)
  if cont>=100 and cont<150:
    an_vir.append(i)
  cont+=1
```

```python
class Osos:
  def __init__(self, lar_set, lar_ver, lar_vir, an_set, an_ver, an_vir):
    self.lar_set=lar_set
    self.lar_ver=lar_ver
    self.lar_vir=lar_vir
    self.an_set=an_set
    self.an_ver=an_ver
    self.an_vir=an_vir
    self.largo=len(lar_ver)

  def ajustar(self, entrenamiento):
    #Regla de  3
    x=100/entrenamiento
    x=self.largo/x
    self.prueva=self.largo-x
    print("prueva",self.prueva)
    #Largos entrenamiento
    en_lar_set=[]
    en_lar_ver=[]
    en_lar_vir=[]
    #Largos prueva
    pr_lar_set=[]
    pr_lar_ver=[]
    pr_lar_vir=[]
    #Anchos entrenamiento
    en_an_set=[]
    en_an_ver=[]
    en_an_vir=[]
    #Anchos prueva
    pr_an_set=[]
    pr_an_ver=[]
    pr_an_vir=[]
    #For chido
    for i in range(self.largo):
      if i<x:
        en_lar_set.append(self.lar_set[i])
        en_lar_ver.append(self.lar_ver[i])
        en_lar_vir.append(self.lar_vir[i])

        en_an_set.append(self.an_set[i])
        en_an_ver.append(self.an_ver[i])
        en_an_vir.append(self.an_vir[i])
      else:
        pr_lar_set.append(self.lar_set[i])
        pr_lar_ver.append(self.lar_ver[i])
        pr_lar_vir.append(self.lar_vir[i])

        pr_an_set.append(self.an_set[i])
        pr_an_ver.append(self.an_ver[i])
        pr_an_vir.append(self.an_vir[i])

    self.en_lar_set=en_lar_set
    self.en_lar_ver=en_lar_ver
    self.en_lar_vir=en_lar_vir

    self.en_an_set=en_an_set
    self.en_an_ver=en_an_ver
    self.en_an_vir=en_an_vir

    self.pr_lar_set=pr_lar_set
    self.pr_lar_ver=pr_lar_ver
    self.pr_lar_vir=pr_lar_vir

    self.pr_an_set=pr_an_set
    self.pr_an_ver=pr_an_ver
    self.pr_an_vir=pr_an_vir



  def fit(self):
    #Centros de clase
    c1=[]
    c2=[]
    c3=[]
    #Contadores de promedio
    l1=0
    l2=0
    l3=0
    a1=0
    a2=0
    a3=0
    #Tamaño
    tama=len(self.en_lar_vir)
    #For chido
    for i in range(tama):
      l1+=self.en_lar_set[i]
      a1+=self.en_an_set[i]

      l2+=self.en_lar_ver[i]
      a2+=self.en_an_ver[i]

      l3+=self.en_lar_vir[i]
      a3+=self.en_an_vir[i]

    cal=l1/tama
    c1.append(cal)
    cal=a1/tama
    c1.append(cal)

    cal=l2/tama
    c2.append(cal)
    cal=a2/tama
    c2.append(cal)

    cal=l3/tama
    c3.append(cal)
    cal=a3/tama
    c3.append(cal)

    print(c1)
    print(c2)
    print(c3)

    self.centro1=c1
    self.centro2=c2
    self.centro3=c3

    return c1,c2,c3


  def predict(self,clases):
    #2 Clases
    if(clases==2):
      desicion=[]
      #Setosa
      for i in range(len(self.pr_lar_set)):
        p1=self.pr_lar_set[i]-self.centro1[0]
        p1=pow(p1,2)
        p2=self.pr_an_set[i]-self.centro1[1]
        p2=pow(p2,2)
        p3=p1+p2

        q1=self.pr_lar_set[i]-self.centro2[0]
        q1=pow(q1,2)
        q2=self.pr_an_set[i]-self.centro2[1]
        q2=pow(q2,2)
        q3=q1+q2

        if(p3<q3):
          desicion.append("set")
        else:
          desicion.append("ver")

      #Versicolor
      for i in range(len(self.pr_lar_ver)):
        p1=self.pr_lar_ver[i]-self.centro1[0]
        p1=pow(p1,2)
        p2=self.pr_an_ver[i]-self.centro1[1]
        p2=pow(p2,2)
        p3=p1+p2

        q1=self.pr_lar_ver[i]-self.centro2[0]
        q1=pow(q1,2)
        q2=self.pr_an_ver[i]-self.centro2[1]
        q2=pow(q2,2)
        q3=q1+q2

        if(p3<q3):
          desicion.append("set")
        else:
          desicion.append("ver")
      print(desicion)
      return desicion

    #3 Clases
    if(clases==3):
      desicion=[]
      #Setosa
      for i in range(len(self.pr_lar_set)):
        p1=self.pr_lar_set[i]-self.centro1[0]
        p1=pow(p1,2)
        p2=self.pr_an_set[i]-self.centro1[1]
        p2=pow(p2,2)
        p3=p1+p2

        q1=self.pr_lar_set[i]-self.centro2[0]
        q1=pow(q1,2)
        q2=self.pr_an_set[i]-self.centro2[1]
        q2=pow(q2,2)
        q3=q1+q2

        r1=self.pr_lar_set[i]-self.centro3[0]
        r1=pow(r1,2)
        r2=self.pr_an_set[i]-self.centro3[1]
        r2=pow(r2,2)
        r3=r1+r2

        if(p3<q3 and p3<r3):
          desicion.append("set")
        elif(q3<p3 and q3<r3):
          desicion.append("ver")
        elif(r3<p3 and r3<q3):
          desicion.append("vir")

      #Versicolor
      for i in range(len(self.pr_lar_ver)):
        p1=self.pr_lar_ver[i]-self.centro1[0]
        p1=pow(p1,2)
        p2=self.pr_an_ver[i]-self.centro1[1]
        p2=pow(p2,2)
        p3=p1+p2

        q1=self.pr_lar_ver[i]-self.centro2[0]
        q1=pow(q1,2)
        q2=self.pr_an_ver[i]-self.centro2[1]
        q2=pow(q2,2)
        q3=q1+q2

        r1=self.pr_lar_ver[i]-self.centro3[0]
        r1=pow(r1,2)
        r2=self.pr_an_ver[i]-self.centro3[1]
        r2=pow(r2,2)
        r3=r1+r2

        if(p3<q3 and p3<r3):
          desicion.append("set")
        elif(q3<p3 and q3<r3):
          desicion.append("ver")
        elif(r3<p3 and r3<q3):
          desicion.append("vir")

      #Virginica
      for i in range(len(self.pr_lar_vir)):
        p1=self.pr_lar_vir[i]-self.centro1[0]
        p1=pow(p1,2)
        p2=self.pr_an_vir[i]-self.centro1[1]
        p2=pow(p2,2)
        p3=p1+p2

        q1=self.pr_lar_vir[i]-self.centro2[0]
        q1=pow(q1,2)
        q2=self.pr_an_vir[i]-self.centro2[1]
        q2=pow(q2,2)
        q3=q1+q2

        r1=self.pr_lar_vir[i]-self.centro3[0]
        r1=pow(r1,2)
        r2=self.pr_an_vir[i]-self.centro3[1]
        r2=pow(r2,2)
        r3=r1+r2

        if(p3<q3 and p3<r3):
          desicion.append("set")
        elif(q3<p3 and q3<r3):
          desicion.append("ver")
        elif(r3<p3 and r3<q3):
          desicion.append("vir")

      print(desicion)
      return desicion

  def matriz(self, clases,mat):
    #2 Clases
    if(clases==2):
      mat2=[]
      rango=int(self.prueva)
      for i in range(rango*clases):
        if(i<rango):
          mat2.append("set")
        else:
          mat2.append("ver")
      mat_conf=confusion_matrix(mat2, mat, labels=["set", "ver"])
      print("La matriz de confucion")
      print(mat_conf)
      print("Acuracy")
      ac=accuracy_score(mat2,mat)
      print(ac)

    #3 Clases
    if(clases==3):
      mat2=[]
      rango=int(self.prueva)
      for i in range(rango*clases):
        if(i<rango):
          mat2.append("set")
        elif(i>=rango and i<(rango*2)):
          mat2.append("ver")
        else:
          mat2.append("vir")
      mat_conf=confusion_matrix(mat2, mat, labels=["set", "ver", "vir"])
      print("La matriz de confucion")
      print(mat_conf)
      print("Acuracy")
      ac=accuracy_score(mat2,mat)
      print(ac)
```
